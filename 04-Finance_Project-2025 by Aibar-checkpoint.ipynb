```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
BAC = data.DataReader("BAC", 'stooq', start, end)
C = data.DataReader("C", "stooq", start, end)
GS = data.DataReader("GS", "stooq", start, end)
JPM = data.DataReader("JPM", "stooq", start, end)
MS = data.DataReader("MS", "stooq", start, end)
WFC = data.DataReader("WFC", "stooq", start, end)
```

```python
display(BAC)
```

```python
df = {
    'BAC': BAC,
    'C': C,
    'GS': GS,
    'JPM': JPM,
    'MS': MS,
    'WFC': WFC
}
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

** Set the column name levels (this is filled out for you):**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for ticker in tickers:
    returns[f'{ticker} Return'] = bank_stocks[ticker]['Close'].pct_change()
returns.head()
```

### Show BAC DataFrame

```python
BAC = bank_stocks['BAC'].sort_index()
returns.head()
```

**The BAC index is not sorted by date.**

```python
BAC = bank_stocks['BAC'].sort_index()
BAC_2008 = BAC.loc['2008-01-01':'2008-12-31']
BAC_2008['MA30'] = BAC_2008['Close'].rolling(window=30).mean()
plt.figure(figsize=(12,6))

plt.plot(BAC_2008['MA30'], label='30 Day Avg')
plt.plot(BAC_2008['Close'], label='BAC CLOSE')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(BAC_2008['MA30'], label='30 Day Avg')
plt.plot(BAC_2008['Close'], label='BAC CLOSE')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
close_prices = bank_stocks.xs('Close', axis=1, level='Stock Info')
close_prices.corr()
```

```python
corr = close_prices.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,fmt=".2f",cmap='viridis',square=True,cbar=True)
plt.xlabel('Bank Ticker')
plt.ylabel('Bank Ticker')
plt.title('Correlation Heatmap of Bank Close Prices')
plt.show()
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(corr,annot=True,fmt=".2f",cmap='rocket',figsize=(8,8))
plt.show()
```

```python
import plotly.express as px
close_prices = bank_stocks.xs('Close', axis=1, level='Stock Info')
corr = close_prices.corr()
fig = px.imshow(corr,aspect="auto",zmin=-1, zmax=1)
fig.update_layout(xaxis_title="",yaxis_title="",)
fig.show()
```
